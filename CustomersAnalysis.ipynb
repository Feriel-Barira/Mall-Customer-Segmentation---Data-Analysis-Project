```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Charger le dataset
url = "Mall_Customers.csv"
data = pd.read_csv(url)
```

```python
print(data.head())
```

```python
# Sélectionner les colonnes pertinentes pour le clustering
data_cleaned = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
print(data_cleaned.head())
```

```python
# Créer une instance de StandardScaler
scaler = StandardScaler()
```

```python
# Appliquer la standardisation sur les données
data_scaled = scaler.fit_transform(data_cleaned)
```

```python
# Convertir les données standardisées en DataFrame
data_scaled_df = pd.DataFrame(data_scaled, columns=data_cleaned.columns)
```

```python
print(data_scaled_df.head())
```

```python
# Appliquer K-Means (par exemple, 4 clusters)
kmeans = KMeans(n_clusters=4, random_state=42)
data['Cluster'] = kmeans.fit_predict(data_scaled)
```

```python
# Visualisation des clusters
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=data, palette='Set1')
plt.title('Segmentation des Clients du Centre Commercial')
plt.show()
```
